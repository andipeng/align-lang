```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
CB91_Green = '#47DBCD'
CB91_Amber = '#F5B14C'
CB91_Blue = '#2CBDFE'
CB91_Pink = '#F3A0F2'
CB91_Purple = '#9D2EC5'
CB91_Violet = '#661D98'
CB91_Amber = '#F5B14C'
color_list = [CB91_Green, CB91_Amber, CB91_Amber, CB91_Violet]
```

```python
#red heart
x = [5, 10, 20, 30, 60]
abstract = [0,0.2,0.55,1,1]
bc = [0,0.1,0.35,0.8,1]

fig1 = plt.figure(figsize=(12,9))
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=color_list)
ax1 = fig1.add_subplot()
ax1.set_title("Task: Red Heart")
ax1.plot(x, abstract, '-', label="Co-Training", marker='o', markersize=8, linewidth=5)
ax1.plot(x, bc, '-', label="BC", marker='o', markersize=8, linewidth=5)

ax1.legend(prop={'size': 25},loc='lower right')
for tick in ax1.get_xticklabels():
    tick.set_rotation(45)
ax1.set_ylabel('Success')
ax1.set_xlabel('# Demonstrations')
for item in ([ax1.title, ax1.xaxis.label, ax1.yaxis.label] +
             ax1.get_xticklabels() + ax1.get_yticklabels()):
    item.set_fontsize(30)
fig1.savefig('q1_red_heart.pdf', bbox_inches='tight')
```
