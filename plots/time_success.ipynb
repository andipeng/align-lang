```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams.update({'font.size': 15})
plt.rc('font',family='serif', serif=['Arial'])
frameon=False
```

```python
human_only_success = [1.00, 1.00, 1.00, 0.87, 0.82, 0.82, 0.92, 1.00, 1.00]
LLM_aide_success = [1.00, 1.00, 1.00, 1.00, 0.90, 0.92, 1.00, 1.00, 1.00]
human_only_time = [0.54, 0.46, 0.39, 0.65, 0.64, 0.48, 0.57, 0.45, 0.52]
LLM_aide_time = [0.10, 0.13, 0.08, 0.11, 0.07, 0.11, 0.09, 0.06, 0.09]
LLM_only_time = [0, 0, 0, 0, 0, 0, 0, 0, 0]
LLM_only_success = [1.00, 1.00, 1.00, 1.00, 0.65, 0.75, 1.00, 1.00, 1.00]
```

```python
plt.scatter(human_only_time, human_only_success, alpha=0.5, label='Human-only')
plt.scatter(LLM_aide_time, LLM_aide_success, alpha=0.5, label='LLM-aided')
plt.scatter(LLM_only_time, LLM_only_success, alpha=0.5, label='LLM-only')

plt.title("Time vs. Success")
plt.xlabel("Time (norm across task)")
plt.ylabel("Success")
plt.legend()

plt.savefig('time_success.pdf', format="pdf", bbox_inches="tight")
plt.show()
```

```python
figure1 = plt.bar(range(len(human_only_success)), human_only_success, yerr=None, align='edge', width=0.4, label='Human-Only')
figure2 = plt.bar(range(len(LLM_aide_success)), LLM_aide_success, yerr=None, align='edge', width=-0.4, label='LLM-Aided (DSP)')
figure3 = plt.bar(range(len(LLM_only_success)), LLM_only_success, yerr=None, align='edge', width=-0.4, label='LLM-Only')

#plt.xticks(np.arange(9), ['red heart', 'heart', 'tiger', 'letter', 'consonant', 'vowel', 'letter', 'alphabet', 'kitchen'])
plt.xticks(np.arange(len(human_only_success)), np.arange(1, len(human_only_success)+1))

plt.legend()

plt.title("Success Rate")
plt.xlabel("Task")
plt.ylabel("Success")
sns.despine()

plt.savefig('success_plot.pdf', format="pdf", bbox_inches="tight")
plt.show()
```
