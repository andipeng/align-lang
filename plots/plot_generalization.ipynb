```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams.update({'font.size': 15})
plt.rc('font',family='serif', serif=['Arial'])
frameon=False
```

```python
# changing color
dsp = [1, 0.95]
dsp_error = [0, 0.05]
gcbc = [0.35, 0.1]
gcbc_error = [0.1, 0.1]

figure1 = plt.bar(range(len(dsp)), dsp, yerr=dsp_error, align='edge', width=0.4, label='DSP (ours)', color='green')
figure2 = plt.bar(range(len(gcbc)), gcbc, yerr=gcbc_error, align='edge', width=-0.4, label='GCBC', color='gray')

#plt.xticks(np.arange(9), ['red heart', 'heart', 'tiger', 'letter', 'consonant', 'vowel', 'letter', 'alphabet', 'kitchen'])
plt.xticks(np.arange(2), ['heart','letter'])

#plt.legend()

plt.title("Color Shift")
plt.xlabel("Task")
plt.ylabel("Success")
sns.despine()

plt.savefig('state_color.pdf', format="pdf", bbox_inches="tight")
plt.show()
```
