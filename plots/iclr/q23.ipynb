```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
plt.rcParams.update({'font.size': 20})
plt.rc('font',family='serif', serif=['Arial'])
frameon=False
CB91_Green = '#47DBCD' 
CB91_Amber = '#F5B14C'
CB91_Blue = '#2CBDFE'
CB91_Pink = '#F3A0F2'
CB91_Purple = '#9D2EC5'
CB91_Violet = '#661D98'
CB91_Amber = '#F5B14C'
matplotlib.rc('font',family='serif', serif=['Palatino'])
sns.set_style('white')

pgf_with_rc_fonts = {"pgf.texsystem": "pdflatex"}
matplotlib.rcParams.update(pgf_with_rc_fonts)
matplotlib.rcParams['text.usetex'] = True

def set_style():
    sns.set(font='serif', font_scale=1.55)
    
   # Make the background white, and specify the
    # specific font family
    sns.set_style("white", {
        "font.family": "serif",
        "font.weight": "normal",
        "font.serif": ["Times", "Palatino", "serif"],
        'axes.facecolor': 'white',
        'lines.markeredgewidth': 1})
```

```python
# distractor shift
lga_s = [1, 0.93, 0.97, 0.8]
lga_s_error = [0.0, 0.02, 0.03, 0.06]
lga = [0.95, 0.95, 0.83, 0.82]
lga_err = [0.03, 0.06, 0.11, 0.07]
gcbc_dart = [0.18, 0.07, 0.18, 0.0]
gcbc_dart_err = [0.12, 0.06, 0.04, 0.00]
lga_l = [0.43, 0.3, 0.35, 0.17]
lga_l_err = [0.11, 0.11, 0.05, 0.02]

set_style()
width = 0.22
x = np.arange(len(lga))
fig, ax = plt.subplots(figsize=(10, 6))
figure1 = ax.bar(x - width, lga_s, yerr=lga_s_error, align='edge', width=width, label='LGA-S', color='orange')
figure2 = ax.bar(x, lga, yerr=lga_err, align='edge', width=width, label='LGA', color='deeppink')
figure3 = ax.bar(x + width, gcbc_dart, yerr=gcbc_dart_err, align='edge', width=width, label='GCBC-DART', color='silver')
figure4 = ax.bar(x + width*2, lga_l, yerr=lga_l_err, align='edge', width=width, label='LGA-L', color='dodgerblue')

#plt.xticks(np.arange(9), ['red heart', 'heart', 'tiger', 'letter', 'consonant', 'vowel', 'letter', 'alphabet', 'kitchen'])
ax.set_xticks(np.arange(4), ['heart (PP)','letter (PP)', 'bowl (rotate)', 'block (sweep, avoid)'])
#ax.legend(loc='right')
ax.set_ylim(0, 1.1)

plt.title("State Covariate Shift")
plt.xlabel("Task")
plt.ylabel("Success")
sns.despine()

plt.savefig('q2.pdf', format="pdf", bbox_inches="tight")
plt.show()
```
