```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams.update({'font.size': 15})
plt.rc('font',family='serif', serif=['Arial'])
frameon=False
CB91_Green = '#47DBCD' 
CB91_Amber = '#F5B14C'
CB91_Blue = '#2CBDFE'
CB91_Pink = '#F3A0F2'
CB91_Purple = '#9D2EC5'
CB91_Violet = '#661D98'
CB91_Amber = '#F5B14C'
```

```python
# distractors
lga = [1, 0.95]
lga_error = [0, 0.05]
gcbc = [0.05, 0]
gcbc_error = [0.05, 0]

figure1 = plt.bar(range(len(lga)), lga, yerr=lga_error, align='edge', width=0.4, label='LGA', color=CB91_Green)
figure2 = plt.bar(range(len(gcbc)), gcbc, yerr=gcbc_error, align='edge', width=-0.4, label='GCBC', color=CB91_Amber)

#plt.xticks(np.arange(9), ['red heart', 'heart', 'tiger', 'letter', 'consonant', 'vowel', 'letter', 'alphabet', 'kitchen'])
plt.xticks(np.arange(2), ['heart','letter'])

plt.legend(loc='right')

plt.title("New Distractors")
plt.xlabel("Task")
plt.ylabel("Success")
sns.despine()

plt.savefig('q2_state_distractor.pdf', format="pdf", bbox_inches="tight")
plt.show()
```

```python
# distractor shift
lga = [1, 0.95, 0.95, 0.8]
lga_error = [0, 0.05, 0.11, 0.24]
lga_abstract_only = [0.95, 0.95, 0.9, 0.85]
lga_abstract_only_error = [0, 0, 0.15, 0.2]
gcbc = [0.15, 0.1, 0.2, 0.0]
gcbc_error = [0.1, 0.1, 0.1, 0.1]
gcbc_abstract_only = [0.45, 0.35, 0.4, 0.15]
gcbc_abstract_only_error = [0.1, 0.2, 0.1, 0.05]

width = 0.22
x = np.arange(len(lga))
fig, ax = plt.subplots(figsize=(10, 6))
figure1 = ax.bar(x - width, lga, yerr=lga_error, align='edge', width=width, label='LGA (abs+state)', color='red')
figure2 = ax.bar(x, lga_abstract_only, yerr=lga_abstract_only_error, align='edge', width=width, label='LGA (abs. only)', color=CB91_Blue)
figure3 = ax.bar(x + width, gcbc, yerr=gcbc_error, align='edge', width=width, label='GCBC (lang)', color=CB91_Amber)
figure4 = ax.bar(x + width*2, gcbc_abstract_only, yerr=gcbc_abstract_only_error, align='edge', width=width, label='GCBC (lang abs.)', color=CB91_Pink)

#plt.xticks(np.arange(9), ['red heart', 'heart', 'tiger', 'letter', 'consonant', 'vowel', 'letter', 'alphabet', 'kitchen'])
ax.set_xticks(np.arange(4), ['heart (PP)','letter (PP)', 'bowl (rotate)', 'block (sweep, avoid)'])
ax.legend(loc='right')
ax.set_ylim(0, 1.1)

plt.title("New Distractors")
plt.xlabel("Task")
plt.ylabel("Success")
sns.despine()

plt.savefig('q2_state_distractor.pdf', format="pdf", bbox_inches="tight")
plt.show()
```

```python
# Multitask
gcbc = [0.2, 0.1, 0]
gcbc_error = [0.15, 0.05,0]
lga = [0.9,1.,0.9]
lga_error = [0.05, 0, 0.1]

figure1 = plt.bar(range(len(lga)), lga, yerr=lga_error, align='edge', width=0.4, label='LGA', color=CB91_Green)
figure2 = plt.bar(range(len(gcbc)), gcbc, yerr=gcbc_error, align='edge', width=-0.4, label='GCBC', color=CB91_Amber)

plt.xticks(np.arange(3), ['bowl','star','find in kitchen'])

plt.legend(loc='right')

plt.title("Multi-Task Ambiguity")
plt.xlabel("Task")
plt.ylabel("Success")
sns.despine()

plt.savefig('q3_lang_ambiguity.pdf', format="pdf", bbox_inches="tight")
plt.show()
```
